# 이미지 읽어오기

```python
# 첫번째 인자: 읽어올 이미지 경로
# 두번째 인자: 옵션 (1=cv.IMREAD_COLOR, 0=cv.IMREAD_GRAYSCALE, -1=cv.IMREAD_UNCHANGED)
# cv.IMREAD_COLOR : 기본세팅, 컬러로 나옴(투명도 무시함)
# cv.IMREAD_UNCHANGED : alpha 채널 포함해서 출력

img = cv.imread("openCV/suzy.jpg", 1)
```

```python
# print 찍어보면 numpy 포맷으로 되어 있음.
# 이미지가 안 불러와졌어도 에러가 발생하지 않음.
# 이미지가 안 불러와졌으면 print 찍었을때 None을 뱉음
# 이미지 경로를 제대로 썼는지 확인.

print(img)
```

```python
# 새창에서 이미지를 띄워줌
# 첫번째 인자: 새로 띄울 창의 이름(임의로 설정함)
# 두번째 인자: 읽어올 이미지 객체

# 새창은 뜨지만 아직 이미지 안나옴

cv.imshow('image', img)
```

```python
# 키보드 이벤트와 연결된 함수.
# 인자는 millisecond 단위로 입력한다.
# 인자로 전달된 밀리세컨드 이내에 아무 키나 입력하면 프로그램이 계속된다.
# 0을 입력하면 무기한으로 키보드 입력을 기다린다.

# 아무키나 누르기 전까지 실행중 * 표시 뜸
# 아무키나 누르면 멈춤.

# 특정 키를 입력할때까지 기다리는 기능도 있으나 나중에 배울 예정.

cv.waitKey(0)
```

```python
# 띄워진 창을 모두 종료함
# 특정 창만 끄려면 cv.destroyWindow("창이름")

cv.destroyAllWindows()
```

```python
# 창을 먼저 띄우고 이미지를 로드할 수도 있다. (위와 반대 순서)
# default값은 cv.WINDOW_AUTOSIZE
# 창 크기를 조절할 수 없다.

cv.namedWindow("image", cv.WINDOW_AUTOSIZE)
```

```python
cv.imshow("image", img)
```

```python
cv.waitKey(10000)
```

```python
cv.destroyWindow("image")
```

```python
# 옵션을 cv.WINDOW_NORMAL로 하면, 크기를 조절할 수 있는 창이 된다.
# 나중에 사이즈가 큰 이미지를 불러올때 유용함.

cv.namedWindow("image", cv.WINDOW_NORMAL)
```

```python
cv.imshow('image', img)
```

```python
cv.waitKey(10000)
```

```python
cv.destroyAllWindows()
```

# 응용

```python
img = cv.imread('openCV/suzy.jpg', 0)
cv.imshow('image', img)
k = cv.waitKey(0) 
# 64비트를 사용할 경우 다음과 같이 하라는데 안해도 잘 되는거 같은데..?
# k = cv.waitKey(0) & 0xFF

if k == 27: # ESC 키를 누르면
    cv.destroyWindow('image')
elif k == ord('s'): # s키를 누르면
    cv.imwrite("openCV/suzy_app.png", img)
    cv.destroyWindow("image")
```

# matplotlib

```python
from matplotlib import pyplot as plt
```

```python
# 이미지 띄우기
# openCV는 BGR 방식으로 이미지를 읽는다.
# matplotlib은 RGB 방식으로 이미지를 읽기 때문에 openCV로 read한 이미지 객체를 컬러로 읽으면 이상하게 보인다.
# 바로잡는 방법은 뒤에서 알려줌.

plt.imshow(img)
```

```python
# 컬러맵 지정하기

plt.imshow(img, cmap='gray')
```

```python
# interpolation(보간법) 지정하기  ->  이미지 크기 바꿀때 적용되는 옵션

plt.imshow(img, cmap='gray', interpolation='bicubic')
```

```python
# x축, y축 스케일 안보이게 하기

plt.xticks([]), plt.yticks([])
```

```python
# 여러 설정 합쳐서 이미지 띄우기

plt.show()
```

```python
# 한방에 실행하면 이렇게 됨.

plt.imshow(img, cmap='gray', interpolation='bicubic')
plt.xticks([]), plt.yticks([])
plt.show()
```

# openCV(BGR)과 matplotlib(RGB) 호환

```python
bgr_img = cv.imread('openCV/suzy.jpg', 1) # 1은 color
```

```python
rgb_img = cv.cvtColor(bgr_img, cv.COLOR_BGR2RGB)
rgb_img
```

```python
# 잘 나옴.

plt.imshow(rgb_img)
```

```python
# 번외
# plt로 읽어들인 이미지를 openCV에서 열면 또 이상함.

img_by_plt = plt.imread("openCV/suzy.jpg")
```

```python
cv.imshow('image',img_by_plt)
cv.waitKey(0)
```

```python
cv.destroyAllWindows()
```

```python
# 번외2
# 이미지는 넘파이 객체로 되어있다.
# 넘파이 재정렬을 통해서 bgr을 rgb로 바꿀 수도 있다.

plt.imshow(bgr_img[:,:,::-1])
```
